```python
some_data = np.random.normal(0,2,size=500)
```

```python
some_data
```

```python
len(str(some_data))
```

```python
plt.plot(some_data)
```

```python
some_data = np.sort(some_data)
plt.hist(some_data,200);#200 is the number of bins
```

```python
np.mean(some_data)
```

```python
np.std(some_data)
```

```python
np.mean([1,2,3,4,5])
```

```python
np.std([1,2,3,4,5])
```

```python
sorted_some_data = np.sort(some_data)
plt.subplot(1,2,1)
plt.xlabel("Independent variable")
plt.ylabel("Dependent variable")
plt.plot(sorted_some_data,label ="Original") #plot dneotes that we have a lot values closer to 0
plt.legend();
plt.subplot(1,2,2)
plt.xlabel("Independent variable")
plt.ylabel("Dependent variable")
plt.plot((sorted_some_data - np.min(sorted_some_data))/(np.max(sorted_some_data)- np.min(sorted_some_data)),label = "normalized data")#We are normalizing our data in the range 1 to -1
plt.legend();
```

```python
np.min(sorted_some_data)
```

```python
sorted_some_data = np.sort(some_data)
plt.subplot(1,2,1)
plt.xlabel("Independent variable")
plt.ylabel("Dependent variable")
noise =np.random.normal(0,1,500)
noisy_data = sorted_some_data+noise
plt.plot(noisy_data,label ="Original") #plot dneotes that we have a lot values closer to 0
plt.legend();
plt.subplot(1,2,2)
plt.xlabel("Independent variable")
plt.ylabel("Dependent variable")
plt.plot((noisy_data - np.min(noisy_data))/(np.max(noisy_data)- np.min(noisy_data)),label = "normalized data")#We are normalizing our data in the range 1 to -1
plt.legend();
```
